```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import random
import pickle
import tensorflow as tf
import keras
```

```python
DIRECTORY=r'C:\JAGA\Jupyter\Dataset'
CATEGORIES=['normal_1','pothole_1']
```

```python
IMG_SIZE=224;
data=[]

for category in CATEGORIES:
    folder=os.path.join(DIRECTORY,category)
    label=CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path=os.path.join(folder,img)
        img_arr=cv2.imread(img_path)
        img_arr=cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
        data.append([img_arr,label])
```

```python
len(data)
```

```python
random.shuffle(data)
```

```python
X=[]
Y=[]

for features,labels in data:
    X.append(features)
    Y.append(labels)
```

```python
X=np.array(X)
Y=np.array(Y)

X=X/255
```

```python
X.shape
```

```python
pickle.dump(data,open("dataset_input",'wb'))
```

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense
```

```python
VGG = tf.keras.applications.VGG16(input_shape=(224,224,3),include_top=False,weights='imagenet')
VGG.trainable = False
```

```python
VGG_model = keras.Sequential([
VGG,
keras.layers.Flatten(),
keras.layers.Dense(units=256,activation="relu"),
keras.layers.Dense(units=256,activation="relu"),
keras.layers.Dense(units=2,activation="softmax")
])
```

```python
VGG_model.compile(optimizer="adam",loss=keras.losses.sparse_categorical_crossentropy,metrics=["accuracy"])
```

```python
VGG_model.fit(X,Y,epochs=10,validation_split=0.2,verbose=1)
```

```python
VGG_model.save("models/VGG_model.h5")
```

```python
predict_data=data[21]
x=predict_data[0]/255
x
```
